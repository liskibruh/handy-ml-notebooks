```python
import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
df = pd.read_csv('Predicted_Gestures.csv')
df.head()
```

```python
df.pop('Unnamed: 0')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df = df.dropna(axis = 0)
```

```python
df.isna().sum()
```

```python
x = df
y = df.pop("Target")
```

```python
y.shape
```

```python
# Perform LDA
lda = LinearDiscriminantAnalysis(n_components=2) # Number of linear 
x_lda = lda.fit_transform(x, y) # Perform LDA and store the transformed data in 
```

```python
# Convert X_lda to a DataFrame
df_lda = pd.DataFrame(x_lda, columns=['LD1', 'LD2'])
```

```python
df_lda
```

```python
df_lda.reset_index(inplace=True)
```

```python
df_lda
```

```python
df_y = pd.DataFrame(y)
df_y
```

```python
df_y.reset_index(inplace = True)
df_y
```

```python
df_y.pop('level_0')
df_y.pop('index')
df_y
```

```python
df_result = pd.DataFrame(x_lda,columns=['LD1','LD2'])
df_result = df_result.assign(target=df_y)
```

```python
df_result
```

```python
df_result.to_csv('LDA_Transformed.csv')
```
