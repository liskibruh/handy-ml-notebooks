```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def deriv_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

def tanh(x):
    return 2 / (1 + np.exp(-2 * x)) - 1

def deriv_tanh(x):
    return 1 - tanh(x) ** 2

def deriv_relu(x):
    return (x > 0) * 1

x = np.arange(-5., 5., 0.2)
fn1 = deriv_tanh(x)
fn2 = deriv_sigmoid(x)
fn3 = deriv_relu(x)

plt.figure(figsize=(8, 6), dpi=80)

plt.plot(x, fn1, label = "function 1")
plt.plot(x, fn2, label = "function 2")
plt.plot(x, fn3, label = "function 3")

plt.legend()
plt.show()
```
